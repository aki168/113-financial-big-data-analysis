```python
import yfinance as yf
import pandas as pd

start_date = "2022-02-01"
end_date = "2024-12-01"

data = yf.download(top_level_code + second_level_code, start=start_date, end=end_date)
closing_prices = data['Adj Close'].dropna(how='all', axis=1)
closing_prices.index = pd.to_datetime(closing_prices.index).strftime('%Y-%m-%d')
```

```python
import numpy as np

n = 50 

def init_population(num, n_stocks,selected_num=15):
  population = []
  for _ in range(num):
    data = np.zeros(n_stocks, dtype=bool)
    indices = np.random.choice(np.arange(n_stocks), size=selected_num, replace=False)
    data[indices] = True
    population.append(data)
  return population

init_group = init_population(n, len(top_level) + len(second_level))
init_group

init_position = pd.DataFrame(init_group, columns=top_level_code + second_level_code)
init_position.columns = pd.Index([f'{col}'[:4] for col in init_position.columns], name='symbol')
init_position
```

```python
returns = closing_prices.pct_change().dropna()
returns.columns = pd.Index([f'{col}'[:4] for col in returns.columns], name='symbol')
df_returns = pd.DataFrame(returns)
df_returns
```

```python
def calculate_sharpe_ratio(returns, risk_free_rate=0.01):
    """計算夏普比率，年化收益率和標準差"""
    mean_return = returns.mean()
    std_return = returns.std()
    sharpe_ratio = (mean_return - risk_free_rate / 252) / std_return * np.sqrt(252) 
    return sharpe_ratio
```

```python
def calculate_max_drawdown(cumulative_returns):
    """計算最大回撤"""
    peak = cumulative_returns.max()
    trough = cumulative_returns[cumulative_returns.idxmax():].min()
    max_drawdown = (trough - peak) / peak
    return max_drawdown
```

```python
def fitness_function(stock, df_returns):
    """計算個體的適應度，股票組合的夏普比率"""
    print(stock)
    selected_stocks = df_returns.columns[stock]
    portfolio_returns = df_returns[selected_stocks].mean(axis=1)  # 計算選定股票的平均日收益率
    return calculate_sharpe_ratio(portfolio_returns)
```

```python
def evaluate_population(population, df_returns, risk_free_rate=0.01, lambda_drawdown=0.5):
    """評估整個種群的適應度"""
    fitness_scores = []
    for stock in population:
        selected_stocks = df_returns.columns[stock]
        portfolio_returns = df_returns[selected_stocks].mean(axis=1)
        cumulative_returns = (1 + portfolio_returns).cumprod()  # 累計收益率

        # 計算夏普比率和最大回撤
        sharpe = calculate_sharpe_ratio(portfolio_returns, risk_free_rate)
        max_drawdown = calculate_max_drawdown(cumulative_returns)

        # 適應度評分，最大回撤作為懲罰項
        fitness = sharpe - lambda_drawdown * np.abs(max_drawdown)
        fitness_scores.append(fitness)
    return fitness_scores

fitness_scores = evaluate_population(init_group, df_returns)
fitness_scores
```

```python
def normalize_fitness_scores(scores):
    min_score = min(scores)
    if min_score < 0:
        scores = [score + abs(min_score) + 0.01 for score in scores]  
    total = sum(scores)
    probabilities = [score / total for score in scores]
    return probabilities

def selection(population, fitness_scores):
    probabilities = np.array(fitness_scores) / np.sum(fitness_scores)
    probabilities = normalize_fitness_scores(probabilities)
    chosen_indices = np.random.choice(len(population), size=len(population), replace=True, p=probabilities)
    selected = [population[i] for i in chosen_indices]
    return selected
```

```python
def crossover(individual1, individual2):
    point = np.random.randint(1, len(individual1))
    child1 = np.concatenate([individual1[:point], individual2[point:]])
    child2 = np.concatenate([individual2[:point], individual1[point:]])
    
    # 確保每個子代都有15個True
    def fix_individual(ind):
        true_indices = np.where(ind)[0]
        false_indices = np.where(~ind)[0]
        if len(true_indices) > 15:
            # 如果True超過15個，隨機將多餘的True改為False
            to_flip = np.random.choice(true_indices, len(true_indices) - 15, replace=False)
            ind[to_flip] = False
        elif len(true_indices) < 15:
            # 如果True不足15個，隨機將False改為True
            to_flip = np.random.choice(false_indices, 15 - len(true_indices), replace=False)
            ind[to_flip] = True
        return ind

    child1 = fix_individual(child1)
    child2 = fix_individual(child2)
    
    return child1, child2
```

```python
def mutate(individual, mutation_rate=0.05):
    for i in range(len(individual)):
        if np.random.rand() < mutation_rate:
            # 隨機選一個位點，與另一個位點交換True或False
            swap_with = np.random.randint(0, len(individual))
            individual[i], individual[swap_with] = individual[swap_with], individual[i]
    return individual
```

```python
def run_genetic_algorithm(df_returns, risk_free_rate, num_generations=50):
    num_stocks = df_returns.shape[1]
    population = init_population(50, num_stocks, 15)  
    for _ in range(num_generations):
        fitness_scores = evaluate_population(population, df_returns, risk_free_rate)  
        population = selection(population, fitness_scores)
        new_population = []
        for i in range(0, len(population), 2):
            if i + 1 < len(population): 
                child1, child2 = crossover(population[i], population[i+1])
                new_population.append(mutate(child1))
                new_population.append(mutate(child2))
        if len(population) % 2 == 1: 
            new_population.append(population[-1])
        population = new_population
    return population
```

```python
result = run_genetic_algorithm(df_returns, 0.01, 50)
```

```python
def calculate_sortino_ratio(returns, risk_free_rate=0.01):
    excess_returns = returns - risk_free_rate / 252
    negative_returns = excess_returns[excess_returns < 0]
    down_std_dev = np.std(negative_returns)
    mean_excess_return = np.mean(excess_returns) * 252
    sortino_ratio = mean_excess_return / down_std_dev if down_std_dev != 0 else 0
    return sortino_ratio
```

```python
# 步驟 5：評估和選擇最優投資組合

# by Sortino Ratio
def select_best(population, df_returns, risk_free_rate):
    best_ratio = float('-inf')
    best_portfolio = None
    
    for individual in population:
        selected_stocks = df_returns.columns[individual]
        portfolio_returns = df_returns[selected_stocks].mean(axis=1)
        sortino_ratio = calculate_sortino_ratio(portfolio_returns, risk_free_rate)
        
        if sortino_ratio > best_ratio:
            best_ratio = sortino_ratio
            best_portfolio = individual

    return best_portfolio
```

```python
window_size = 90 

historical_positions = df_returns.copy()
historical_positions = historical_positions.astype(bool)

for end_date in df_returns.index[window_size:]:
    start_date = pd.to_datetime(end_date) - pd.DateOffset(days=window_size)
    start_date = start_date.strftime('%Y-%m-%d')
    window_returns = df_returns.loc[start_date:end_date]

    population_r = run_genetic_algorithm(window_returns, risk_free_rate=0.01, num_generations=50)
    
    best_population = select_best(population_r, window_returns, risk_free_rate=0.01)
    
    historical_positions.loc[end_date] = best_population
```
